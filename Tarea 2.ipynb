```python
#paqueterias
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt

#variables
x=sy.symbols('x')

#funciones
f=sy.sin(x)+sy.cos(x)
ff=sy.lambdify(x,f)

#aproximacion
g=sy.series(f,x,0,15)
gg=sy.lambdify(x,g.removeO())

#error
e=ff(5)-gg(5)

#tolerancia
t=0.05

#orden
o=g.getO()


#prints
print('La función es')
display(f)
print('La serie de Taylor es:')
display(g)
print('La funcion evaluada en 5 es ')
display (ff(5))
print('La serie evaluada en 5 es')
display (gg(5))
print('El error es ')
display(e)

#restriccion
if abs(e) > t:
    print ("el error es muy grande")
else:
    print('error válido')


#plots
X=np.linspace(0,5)
Yf=ff(X)
Yg=gg(X)
fig,ax=plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.plot(X,Yf)
plt.plot(X,Yg)
plt.legend(['Funcion f','S-T de f'])
plt.show()
print ("El orden de la aproximacion es:", o)
```

```python
#definimos subintervalos
f = sy.sin(x)+sy.cos(x)
g = sy.series(f,x,-1.5,5)
h = sy.series(f,x,-0.5,5)
i = sy.series(f,x,0.5,5)
j = sy.series(f,x,1.5,5)
#quitar errores
gx= g.removeO()
hx= h.removeO()
ix= i.removeO()
jx= j.removeO()
#transformarlas
flam = sy.lambdify(x,f)
glam = sy.lambdify(x,gx)
hlam = sy.lambdify(x,hx)
ilam = sy.lambdify(x,ix)
jlam = sy.lambdify(x,jx)
#evaluar y prints
print(flam(-1.7))
print(glam(-1.7))
print(flam(-0.2))
print(hlam(-0.2))
print(flam(0.9))
print(ilam(0.9))
print(flam(1.8))
print(jlam(1.8))
#errores de cada aprox
eabsg = np.abs(flam(-1.7) - glam(-1.7))
eabsh = np.abs(flam(-0.9) - hlam(-0.9))
eabsi = np.abs(flam(0.9) - ilam(0.9))
eabsj = np.abs(flam(1.8) - jlam(1.8))
gerel = np.abs((eabsg/flam(-1.7))*100)
herel = np.abs((eabsh/flam(-0.9))*100)
ierel = np.abs((eabsi/flam(0.9))*100)
jerel = np.abs((eabsj/flam(1.8))*100)
#prints errores
print(eabsg)
print(gerel,"%")
print(eabsh)
print(herel,"%")
print(eabsi)
print(ierel,"%")
print(eabsj)
print(jerel,"%")
#primer intervalo
X = np.linspace(-2,-1)
g = sy.series(f,x,-1.5,5)
gx = g.removeO()
glam = sy.lambdify(x,gx)
YF = flam(X)
YG = glam(X)
Yerr = np.abs(YF-YG)
Yord = X**5
plt.plot(X,Yerr, label = "Error", color = "Blue")
plt.plot(X,Yord, label = "Orden de la función O", color = "Yellow")
plt.legend(loc=2)
plt.xlabel("X")
plt.ylabel("Error")
plt.title("Gráfica del error")
```
